```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# On lit le fichier excel
dataframe = pd.read_excel('Players_WC2014.xlsx')
# Au cas où, on supprime les colonnes dont les valeurs de chaque chaque ligne est NaN
dataframe = dataframe.dropna(axis = 'columns', how = 'all')
# On retourne le début de la collection
dataframe.head()
```

```python
dataframe.sort_values(by="Distance Covered", ascending=False)[:3]
```

```python
# On recupere le dataframe
# On tri par "Top speed"
# on enleve les lignes ayant NaN

_dataframe = dataframe\
.sort_values(by=["Top Speed"], ascending=False)\
.dropna(subset=['Top Speed'], axis="index")

_dataframe
```

```python
dataframe["Top Speed"] = dataframe["Top Speed"].str.replace("km/h","").astype(float)
dataframe["Distance Covered Not In Possession"] = dataframe["Distance Covered Not In Possession"].str.replace("km/h","").astype(float)
dataframe
```

```python
dataframe["Top Speed"].quantile(q=0.1)
```

```python
decile = math.trunc(dataframe["Top Speed"].quantile(q=0.1))

dataframe[:decile]\
.sort_values(by=["Distance Covered Not In Possession"], ascending=False)
```

```python
subset = dataframe [dataframe["Distance Covered Not In Possession"] > decile]
subset
```
