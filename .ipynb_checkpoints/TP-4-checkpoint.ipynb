```python
dataframe["Top Speed"] = dataframe["Top Speed"].str.replace("km/h","").astype(float)
dataframe["Distance Covered Not In Possession"] = dataframe["Distance Covered Not In Possession"].str.replace("km","").astype(float)
#dataframe
```

```python
decile = math.trunc(dataframe["Top Speed"].quantile(q=0.1))
decile
#dataframe[:decile]\
#.sort_values(by=["Distance Covered Not In Possession"], ascending=False)
```

```python
subset = dataframe [dataframe["Top Speed"] > decile]\
.sort_values(by="Distance Covered Not In Possession", axis="index", ascending=False)[:1]

subset[['Name', 'Team']]
```
