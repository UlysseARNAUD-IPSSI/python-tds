```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def read_csv(file):
    return pd\
        .read_csv(file + '.csv', error_bad_lines=False, index_col=False, dtype='unicode')\
        .dropna(axis = 'columns', how = 'all')
```

```python
def communes(*,metropolitain=True,outremer=False,corse=True):
    
    dfCommunes = read_csv('communes-departement-region')
    
    dfCommunesGeo = dfCommunes[['code_commune_INSEE', 'code_departement', 'longitude', 'latitude']]\
        .dropna(axis='columns',how='all')\
        .dropna(axis='index',how='any',subset=['longitude','latitude'])
    
    dfCommunesGeo.longitude = dfCommunesGeo.longitude.apply(lambda x: "0" if x.strip() == "-" else x).astype(float)
    dfCommunesGeo.latitude = dfCommunesGeo.latitude.astype(float)
    
    subsetFinal = pd.Series(dtype='object', name="Communes")
    
    if True == metropolitain:
        subsetMetropolitain = dfCommunesGeo.loc[\
             (dfCommunesGeo['code_departement'].str.isnumeric())\
             & (dfCommunesGeo['code_departement'].str.len() < 3)
        ]
        subsetFinal['metropolitain'] = subsetMetropolitain
    
    if True == corse:
        subsetCorse = dfCommunesGeo.loc[\
            (dfCommunesGeo['code_departement'] == '2A')\
            | (dfCommunesGeo['code_departement'] == '2B')\
        ]
        subsetFinal['corse'] = subsetCorse
        
    
    if True == outremer:
        subsetOutremer = dfCommunesGeo.loc[\
             (dfCommunesGeo['code_departement'].str.len() > 2)
        ]
        subsetFinal['outremer'] = subsetOutremer
        
    subsetFinal = pd.concat(subsetFinal.to_list(), axis='index').reset_index()
        
    return subsetFinal
```

```python
communes()
```

```python
_communes = communes()
plt.figure(figsize=(10,10))
plt.scatter(
    _communes.longitude, 
    _communes.latitude, 
    s=1
)
plt.colorbar()
```

```python
dfInseeCommunes = read_csv('Data_insee_communes')
# On remplace les noms des colonnes
dfInseeCommunes.rename(columns={
    'CODGEO':'code_commune_INSEE',
    'Nb Pharmacies et parfumerie': 'nombre_pharmacies_parfumeries'
},inplace=True)
# On enleve les premiers 0 présents dans le code géographique
dfInseeCommunes['code_commune_INSEE'] = dfInseeCommunes.apply(lambda x: x.str.strip('0'))
```

## Nb pharmacies et parfumerie 

```python
dfCommunes = read_csv('communes-departement-region')

dfCommunesGeo = dfCommunes[['code_commune_INSEE', 'code_departement', 'longitude', 'latitude']]\
.dropna(axis='columns',how='all')\
.dropna(axis='index',how='any',subset=['longitude','latitude'])

dfCommunesGeo.longitude = dfCommunesGeo.longitude.apply(lambda x: "0" if x.strip() == "-" else x).astype(float)
dfCommunesGeo.latitude = dfCommunesGeo.latitude.astype(float)

dfCommunesGeo['departement'] = dfCommunesGeo.code_commune_INSEE.apply(lambda x: x[:2])

subset = dfCommunesGeo[dfCommunesGeo['departement'] != '97']

subsetOutremer = dfCommunesGeo[dfCommunesGeo['departement'] == '97']
```

```python
dfNbPharmacieParfumerie = dfInseeCommunes[['code_commune_INSEE', 'nombre_pharmacies_parfumeries']]
subset = subset.join(
    dfNbPharmacieParfumerie.set_index('code_commune_INSEE'), 
    on='code_commune_INSEE'
)
subset.dropna(axis='index',how='any',subset=['nombre_pharmacies_parfumeries'])
subset['nombre_pharmacies_parfumeries'] = subset['nombre_pharmacies_parfumeries'].astype(float)
```
