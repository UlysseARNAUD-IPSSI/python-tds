```python
# example of a bimodal data sample
from matplotlib import pyplot
from numpy.random import normal
from numpy import hstack
# generate a sample
sample1 = normal(loc=200, scale=5, size=300)
sample2 = normal(loc=40, scale=50, size=7000)
sample3 = normal(loc=-100, scale=5, size=70)
sample = hstack((sample1, sample2))
sample = hstack((sample, sample3))
sample = list(sample)
```

```python
"""
Donne la moyenne tronquée
Source : https://stackoverflow.com/a/60318236
"""
def tmean(aList, alpha):
    sortedList = sorted(aList)
    
    a = round(alpha * len(aList))
    
    if a == 0:
        return sum(aList) / len(aList)
    
    trimmed_list = sortedList[a:-a]
    
    if len(trimmed_list) == 0:
        return

    return sum(trimmed_list)/len(trimmed_list)
```
